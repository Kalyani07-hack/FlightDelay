Data Cleaning

```python
df.columns = (
    df.columns
    .str.lower()
    .str.strip()
    .str.replace(" ","_")
)
df.columns
```

```python
datetime_cols = [
    'scheduleddeparture', 
    'actualdeparture', 
    'scheduledarrival',
    'actualarrival'
]

for cols in datetime_cols:
    df[cols] = pd.to_datetime(df[cols])
```

```python
df["cancelled"] = df["cancelled"].astype(bool)
df["diverted"] = df["diverted"].astype(bool)
```

```python
df["Early_flag"] = df["delayminutes"] < 0   #boolean vaule will be stored in this new col to indicate early flights
df.loc[df["delayminutes"] < 0, "delayminutes"] = 0   #negative delay are set to 0
df["delayreason"] = df["delayreason"].fillna("No Delay")   #null changed to no delay
```

```python
df["origin"] = df["origin"].str.upper().str.strip()
df["destination"] = df["destination"].str.upper().str.strip()
```

```python
df = df[df["distance"] > 0] #distance should not be zero or negative

df = df[df["flightnumber"].notna()]   #select rows where flight no is not null
```

SQL Connection

```python
from urllib.parse import quote_plus
from sqlalchemy import create_engine

password = quote_plus("Sharmila@07")

engine = create_engine(
    f"mysql+mysqlconnector://root:{password}@localhost:3306/flight_db",
    pool_pre_ping=True,
    pool_recycle=3600
)

# chunked upload
df.to_sql(
    "flights_cleaned",
    con=engine,
    if_exists="replace",
    index=False,
    chunksize=10000,     # 10k rows per batch
    method="multi"       # bulk insert
)
```
